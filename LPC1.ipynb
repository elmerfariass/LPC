```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Vamos usar uma frequncia de amostragem alta para um grfico suave
fs = 1000
# Definimos um tempo de -2 a 4 segundos para lidar bem com os atrasos
# (x(t-1) precisa de valores de t<0 para ser calculado)
t = np.arange(0, 4, 1/fs)

# Sinal de entrada: x(t) = cos(2*pi*t) + (1/3)*cos(6*pi*t)
# (Frequncias: 1 Hz e 3 Hz)
x = np.cos(2 * np.pi * 1 * t) + (1/3) * np.cos(6 * np.pi * 1 * t)


# --- 2. Implementao dos Sistemas ---

# a. y(t) = 2x(t - 1)
# Substitumos 't' por '(t-1)' na frmula original de x(t)
y_a = 2 * (np.cos(2 * np.pi * (t - 1)) + (1/3) * np.cos(6 * np.pi * (t - 1)))

# b. y(t) = x(t) + 0.5x(t - 0.8)
# Precisamos de x(t) (que j temos) e x(t - 0.8)
x_shifted_b = np.cos(2 * np.pi * (t - 0.8)) + (1/3) * np.cos(6 * np.pi * (t - 0.8))
y_b = x + 0.5 * x_shifted_b

# c. y(t) = x^2(t)
y_c = x**2


# --- 3. Plotagem dos Grficos ---
fig, axs = plt.subplots(3, 1, figsize=(8, 8))
fig.suptitle('Anlise de Distoro de Sistemas', fontsize=18)



# --- Grfico (a) ---
axs[0].plot(t, x, label='x(t) [Entrada]', linestyle=':')
axs[0].plot(t, y_a, label='y(t) = 2x(t-1)')
axs[0].set_title('a. Sistema y(t) = 2x(t - 1)')
axs[2].set_xlabel('Tempo (s)')
axs[0].set_ylabel('Amplitude')
axs[0].legend()
axs[0].grid(True)

# --- Grfico (b) ---
axs[1].plot(t, x, label='x(t) [Entrada]', linestyle=':')
axs[1].plot(t, y_b, label='y(t) = x(t) + 0.5x(t-0.8)', color ='red')
axs[1].set_title('b. Sistema y(t) = x(t) + 0.5x(t - 0.8)')
axs[2].set_xlabel('Tempo (s)')
axs[1].set_ylabel('Amplitude')
axs[1].legend()
axs[1].grid(True)

# --- Grfico (c) ---
axs[2].plot(t, x, label='x(t) [Entrada]', linestyle=':')
axs[2].plot(t, y_c, label='y(t) = x(t)',color ='green')
axs[2].set_title('c. Sistema y(t) = x(t)')
axs[2].set_xlabel('Tempo (s)')
axs[2].set_ylabel('Amplitude')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
